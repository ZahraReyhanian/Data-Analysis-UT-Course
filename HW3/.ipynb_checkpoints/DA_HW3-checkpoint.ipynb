4. Most frequent drop-zones

`np.unique` that I explained in the previous part, with parameter `return_counts=True` claculates the number of each unique value and returns them.

I convert the outputs of it to DataFrame to have better calculation.

First, I convert the type of count column to integer and then sort the dataframe I created by this column.

Finally, I returned the last row of previous output which is the most frequent drop zone.

````python
unique, counts = np.unique(list(df_taxis['dropoff_zone']), return_counts=True)
u_dropoff = pd.DataFrame(np.asarray((unique, counts)).T, columns = ['dropoff_zone','count'])
u_dropoff = u_dropoff.astype({'count': 'int32'})
u_dropoff = u_dropoff.sort_values(by='count', axis=0).reset_index().drop(["index"], axis=1)
u_dropoff.tail(1)
````

I get 5 least frequent dropoff zone and 5 most frequent dropoff zone using `head(5)` and `tail(5)`

````python
head = u_dropoff.head(5)
tail = u_dropoff.tail(5)
````

````python
plt.barh(tail['dropoff_zone'], tail['count'])
plt.xlabel('frequency')
plt.ylabel('dropoff zone')
plt.title("5 most frequent dropoff zone")
plt.show()
````

````python
plt.barh(head['dropoff_zone'], head['count'])
plt.xlabel('frequency')
plt.ylabel('dropoff zone')
plt.title("5 least frequent dropoff zone")
plt.show()
````

#### Analysis of two plots above

As you see, `Upper East Side North` is the most frequent droppoff zone with near 250 trips. After it `Murry Hill`, `Midtown Center`, `Upper East Side South`, and `Midtown East` are the most frequent droppoff zone respectively.

There is only one trip to the least frequent droppoff zone. 5 of them were showed in the plot which are `Stapleton`, `Madison`, `Douglaston`, `Brooklyn Navy Yard`, and `Oakland Gardens`.

5. Percentage of drop-offs and pick-ups per weekday

First need to convert the datatype of `pickup` and `dropoff` to `datetime`

Then get day of week of each one.

````python
df_taxis['pickup'] = pd.to_datetime(df_taxis['pickup'])
df_taxis['dropoff'] = pd.to_datetime(df_taxis['dropoff'])
df_taxis['pickup_day']=df_taxis['pickup'].dt.dayofweek
df_taxis['dropoff_day']=df_taxis['dropoff'].dt.dayofweek
````

calculate percentage

````python
pickup = df_taxis[['pickup', 'pickup_day']].groupby(by='pickup_day').count() * 100 / df_taxis.shape[0]
dropoff = df_taxis[['dropoff', 'dropoff_day']].groupby(by='dropoff_day').count() * 100 / df_taxis.shape[0]
````

````python
plt.figure(figsize=(10, 6))
plt.xlabel('day of week')
plt.ylabel('percentage of drop-off and pick-up')
plt.title('Percentage of drop-offs and pick-ups per weekday')

ind = pickup.index

width = 0.4

plt.bar(pickup.index , pickup.pickup, 0.4, label = 'pickup')
plt.bar(dropoff.index + width , dropoff.dropoff, 0.4, label = 'dropoff')

plt.xticks(ind + width / 2, ('Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'))

plt.legend()
plt.show()
````

#### Analysis of graph

The graph above, show that percentages of dropoff and pickup in each day of week are almost equal. A little different must be because of using taxi in midnight, so maybe pickup was at before 24 o'clock of one day and dropoff was after 0 o'clock of the next day.

As it shows, in the Friday and Saturday, using taxi is more than the other days.

Here split passengers by ranges of trips count(at most 200, between 200 and 500, and at least 500)

Finally, created a dataframe based on it.

````python
plt.bar(df_cnt_pass.index, df_cnt_pass['count'])
plt.xlabel('number of trips')
plt.ylabel('frequency')
plt.yticks(np.arange(4))

plt.title('Ranges frequency of passengers trip count')

plt.show()
````

## Part2 - Dimention Rednction

### Selection

We want to choose some features and drop the others. Certainly fare doesn't depend on color of taxi or type of payment. So I deleted them. I also pickup and dropoff datetime and used day of week and hour of them which may effect fare. I also dropped zone and borough of pickup and dropoff and index which don't effect fare.

For the other features, we should know correlation between them and fare. So, I calculated correlation and showed it:

````python
numerical_variables = ['passengers', 'distance', 'fare', 'tip', 'tolls', 'total', 'pickup_day', 'dropoff_day', 'pickup_hour', 'dropoff_hour']
corr = df_taxis[numerical_variables].corr()

f, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), cmap=sns.diverging_palette(220, 10, as_cmap=True), square=True, ax=ax);
````

#### Analysis

First we need to set an absolute value, say 0.6 as the threshold for selecting the variables. If we find that the predictor variables are correlated among themselves, we can drop the variable which has a lower correlation coefficient value with the target variable.

We can see row or column of fare and compare correlation between it and the other features. As you see, the correlation between it and passengers, dropoff_day, pickup_hour, and dropoff_hour is too low and we can delete them easily.

The correlation between fare and tip is less than the threshold and this can be deleted.

As heatmap shows, the correlation coefficient value of other features which are distance, tolls, and total are greater than the threshold. They also have correlation between themselves which means we must choose one of them and drop the others. I keep the highest correlation coefficient value means total and drop the others.

So, I choose `total`.

### PCA

````python
df = df_taxis[['passengers', 'distance', 'tip', 'tolls', 'total', 'color', 'payment', 'pickup_zone', 'dropoff_zone', 'pickup_borough', 'dropoff_borough', 'pickup_hour', 'dropoff_hour']]
````

convert non numeric data to numeric data

````python
df['payment'] = pd.factorize(df['payment'])[0]
df['color'] = pd.factorize(df['color'])[0]
df['pickup_zone'] = pd.factorize(df['pickup_zone'])[0]
df['dropoff_zone'] = pd.factorize(df['dropoff_zone'])[0]
df['pickup_borough'] = pd.factorize(df['pickup_borough'])[0]
df['dropoff_borough'] = pd.factorize(df['dropoff_borough'])[0]
df['pickup_year'] =  df_taxis['pickup'].dt.year
df['pickup_month'] =  df_taxis['pickup'].dt.month
df['pickup_day'] =  df_taxis['pickup'].dt.day
df['dropoff_year'] =  df_taxis['dropoff'].dt.year
df['dropoff_month'] =  df_taxis['dropoff'].dt.month
df['dropoff_day'] =  df_taxis['dropoff'].dt.day
````

````python
# PCA
````

First, we need data that are centered. So I used `StandardScaler` and the method `fit_transform`.The Standard Scaler is a simple transformation that normalizes the data to have a mean of 0 and 0 unit variance, or a standard deviation of 1.

Then using pca, convert the dataset to two-dimensional dataset and then visualize it by `plotly` library:

````python
import plotly.express as px 
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X = scaler.fit_transform(df)
pca = PCA(n_components=2) 
components = pca.fit_transform(X) 
fig = px.scatter(components, x=0, y=1, color=df_taxis['fare']) 
fig.show()
````

````python
# Plot
````

Here we find the explained variance for each principal component. We calculate explained variance by first scaling our data with the StandardScalar and then fitting a PCA model to the scaled data.

You will also notice that the total number of Principal Components equals the number of Features in your dataset. However, one important thing to note is that the Principal Components are not the dataset's columns; rather, PCA is constructing new features that best explain the data.

After calculating cumulative variance, I plot it to show how it grows. Additionally, I plot a line that represents 95% of the explained variance. While there is no rule as to how much explained variance you need to include in a model, 95% is a good threshold to start. 

````python
import plotly.express as px 
import numpy as np 
import pandas as pd 
from sklearn.decomposition import PCA 
from sklearn.datasets import load_diabetes 

pca = PCA() 
pca.fit(X)
exp_var_cumul = np.cumsum(pca.explained_variance_ratio_) 

fig = px.line( x=range(1, exp_var_cumul.shape[0] + 1), y=exp_var_cumul, labels={"x": "# Components", "y": "Explained Variance"} )
fig.add_hline(y = 0.95,
                annotation_text='95% cut-off threshold',
                annotation_position="top left",
                line=dict(color='Red',),    
)
````

After plotting the cumulative explained variance, we can see how the curve flattens slightly around 8 components. And where our line is drawn for 95%, the total explained variance is at approximately 12 components. So we select 12 as the best number of components for pca.

#### Correlation matrix

A correlation matrix is simply a table which displays the correlation coefficients for different variables. The matrix depicts the correlation between all the possible pairs of values in a table.
A correlation matrix consists of rows and columns that show the variables. Each cell in a table contains the correlation coefficient.
Correlation is a measure of the linear relationship of 2 or more variables. Through correlation, we can predict one variable from the other. The logic behind using correlation for feature selection is that the good variables are highly correlated with the target. Furthermore, variables should be correlated with the target but should be uncorrelated among themselves.
If two variables are correlated, we can predict one from the other. Therefore, if two features are correlated, the model only really needs one of them, as the second one does not add additional information.

#### Principal Component Analysis(PCA)

The principal component analysis is a statistical method that allows you to summarise information into smaller sets that may be displayed and studied more readily. Large datasets are becoming more prevalent, yet they might be challenging to analyze. PCA is a technique for decreasing the dimensionality of datasets while enhancing interpretability and avoiding information loss. In the disciplines of statistics and data science, principal component analysis (PCA) is a crucial method to grasp.

##### advantages and disadvantages of correlation matrix

In the first part, I used correlation matrix between features to choose features. 

This way is simple .It enhances the visualization and performance of the algorithm. It doesn't need any preprocessing. We can easily compare the the correlation coefficient value of each variable with others. But it isn't so accurate. Because we lost lots of information.

##### advantages of PCA
- Correlated features are removed

Running algorithms on large datasets with all of the features will lower the speed of your method and make it difficult to show the many characteristics in any type of graph. Finding connections in hundreds of characteristics manually is virtually difficult, tedious, and time-consuming because principal components are independent of one another, and associated characteristics are removed.

- Enhances the performance of the algorithm.

With so many characteristics, your algorithms performance will surge greatly. The principal component analysis is a popular method for speeding up your machine Learning algorithm by removing associated variables that dont help decision-making. With fewer features, the training time of the algorithms decreases considerably. So, if the input dimensions are too large, utilizing PCA to speed up the method is a viable option.

- Enhanced Visualization

PCA converts high-dimensional data to low-dimensional data that can be readily viewed. PCA produces large variance, which helps visualization. PCA is based on linear algebra, which is computationally simple for computers to solve. It accelerates other machine learning methods, allowing them to converge quicker when trained on main components rather than the original dataset.

- Reduces Overfitting

Overfitting mainly occurs when there are too many variables in the dataset. So, PCA helps in overcoming the overfitting issue by reducing the number of features.

##### disadvantages of PCA
- The major components are difficult to comprehend.

After applying principal component analysis to the dataset, your original features will be transformed into Principal Components. Original features are more legible and interpretable than Principal Components. Even the most basic invariance could not be caught by the PCA unless the training data clearly stated it. For example, after computing the main components, it is difficult to determine which characteristics in the dataset are the most significant.

- Data normalization is required.

Before applying principal component analysis, you must normalize your data; otherwise, PCA will be difficult to discover optimal principal components. The scale has an effect on PCA, thus you must scale the features in your data before using PCA. As a result, principal components will be skewed toward characteristics with large variance, resulting in incorrect findings. 

- Loss of information.

PCA accounts for the greatest amount of variation across data characteristics. If the number of Principal Components is not carefully chosen, it may miss certain information in contrast to the real list of characteristics. Although dimensionality reduction is beneficial, it has a cost. Loss of information is an inevitable component of principal component analysis. Managing the trade-off between dimensionality reduction and information loss is, regrettably, an unavoidable tradeoff when employing PCA.

##### comparing the result

Using correlation matrix, we extract one feature that is related to column of fare. We lost some information here.

But pca, extracts 12 principal component with a rather good variance that are different from original features. Principal Components are the linear combination of your original features. Principal Components are not as readable and interpretable as original features. But this way helps not to lose information and have better performace in modeling. This way is more complicated than using correlation matrix but it is more accurate.
